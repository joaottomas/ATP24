```python
import random

def cria_lista():
    n=int(input("Quantos elementos deseja na sua lista?"))
    return [random.randint(1,100) for num in range(n)]

def ler_lista():
    n=int(input("Quantos elementos deseja na sua lista?"))
    i=1
    res =[]
    while i <= n:
        num=int(input("Escreva o numero que voce deseja!"))
        i=i+1
        res.append(num)
    return res

def soma(lista):
    soma=0
    for elem in lista:
        soma=soma+elem
    return soma
    

def media(lista):
     if len(lista) == 0:
        return 0
     som = soma(lista)
     return som / len(lista)
    
def maior(lista):
    maior=lista[0]
    for elem in lista[1:]:
        if elem > maior:
            maior=elem
    return maior

def menor(lista):
    menor=lista[0]
    for elem in lista[1:]:
        if elem < menor:
            menor=elem
    return menor
    


def estaOrdenada_cres(lista):
     i=1
     cond=True
     while i<len(lista) and cond:
        if lista[i] < lista [i-1]:
            cond=False
        i=i+1
     return cond
    


def estaOrdenada_decres(lista):
     i=1
     cond=True
     while i<len(lista) and cond:
        if lista[i] > lista [i-1]:
            cond=False
        i=i+1
     return cond
    

def procurar_elem(lista):
    num=int(input("Qual o numero que deseja procurar na lista?"))
    if num in lista:
        print("O seu numero esta na posição:", lista.index(num))
    else:
        return print("Elemento não esta na lista: -1")
    
def menu():
    print("Menu:")
    print("(1) Criar Lista")
    print("(2) Ler Lista")
    print("(3) Soma")
    print("(4) Média")
    print("(5) Maior")
    print("(6) Menor")
    print("(7) estaOrdenada por ordem crescente")
    print("(8) estaOrdenada por ordem decrescente")
    print("(9) Procura um elemento")
    print("(0) Sair")

def salvar_lista(lista):
    f = open("./ficheiros/memoria.txt",'w')
    for elem in lista:
        f.write(f"{elem}\n")


def carregar_lista():
    f = open("./ficheiros/memoria.txt")
    return [int(linha.strip()) for linha in f]



cond=True
while cond:
    lista = carregar_lista()
    menu()
    opcao=int(input("Escolha uma opção:"))
    if opcao == 1:
        lista = cria_lista()
        salvar_lista(lista)
        print(f"\nLista criada:{lista}\n")
    elif opcao==2:
        lista=ler_lista()
        salvar_lista(lista)
        print(f"\nLista criada:{lista}\n")
    elif opcao==3:
        print(f"\nSoma da lista:{soma(lista)}\n")
    elif opcao == 4:
        print(f"\nMédia da lista:{media(lista)}\n")
    elif opcao == 5:
        print(f"\nMaior elemento da lista:{maior(lista)}\n")
    elif opcao == 6:
        print(f"\nMenor elemento da lista:{menor(lista)}\n")
    elif opcao == 7:
        print(f"\nLista está ordenada por ordem crescente? \n{"Sim" if estaOrdenada_cres(lista) else "Não"}\n")
    elif opcao == 8:
        print(f"\nLista está ordenada por ordem decrescente? \n{"Sim" if estaOrdenada_decres(lista) else "Não"}\n")
    elif opcao == 9:
        procurar_elem(lista)
    else:
        print(f"\nA fechar o programa.....\nLista Final:{lista}\n")
        cond=False
```
