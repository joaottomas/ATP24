```python
def listar(cinema):
    print("---------------Filmes em exibição-----------------")
    for sala in cinema:
        nlugares, ocupados, nome = sala
        print(f"- {nome}")
    print("-------------------------------------------------")
    return

def disponivel(cinema, filme, lugar):
    for sala in cinema:
        nlugares, ocupados, nome = sala
        if nome == filme and lugar <= nlugares:
            if lugar not in ocupados:
                return True
    return False

def vendebilhete(cinema, filme, lugar):
    if disponivel(cinema, filme, lugar):
        for sala in cinema:
            nlugares, ocupados, nome = sala
            if nome == filme:
                ocupados.append(lugar)
    return cinema

def listardisponibilidades(cinema):
    print("Disponibilidade nas salas:")
    for sala in cinema:
        nlugares, ocupados, nome = sala
        lugares_disponiveis = nlugares - len(ocupados)
        print(f"Filme: {nome} - Lugares disponíveis: {lugares_disponiveis}")

def inserirSala(cinema, sala):
    for s in cinema:
        if s[2] == sala[2]:
            print(f"Sala para o filme '{sala[2]}' já existe.")
            return cinema
    cinema.append(sala)
    return cinema

def menu():
    cinema = []
    opcao = ""
    while opcao != "0":
        print("\nMenu:")
        print("1. Listar filmes em exibição")
        print("2. Verificar disponibilidade de um lugar")
        print("3. Vender bilhete")
        print("4. Listar disponibilidades das salas")
        print("5. Inserir nova sala")
        print("0. Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            listar(cinema)

        elif opcao == "2":
            filme = input("Informe o nome do filme: ")
            lugar = int(input("Informe o número do lugar: "))
            if disponivel(cinema, filme, lugar):
                print("Lugar disponível.")
            else:
                print("Lugar já ocupado ou filme não encontrado.")
        
        elif opcao == "3":
            filme = input("Informe o nome do filme: ")
            lugar = int(input("Informe o número do lugar: "))
            cinema = vendebilhete(cinema, filme, lugar)
        
        elif opcao == "4":
            listardisponibilidades(cinema)
        
        elif opcao == "5":
            nlugares = int(input("Informe o número de lugares da sala: "))
            filme = input("Informe o nome do filme: ")
            nova_sala = (nlugares, [], filme)
            cinema = inserirSala(cinema, nova_sala)
        
        elif opcao == "0":
            print("Saindo...")
        
        else:
            print("Opção inválida. Tente novamente.")

# Testar o funcionamento
sala1 = (150, [], "Twilight")
sala2 = (200, [], "Hannibal")
cinema1 = []
cinema1 = inserirSala(cinema1, sala1)
cinema1 = inserirSala(cinema1, sala2)
listar(cinema1)

if disponivel(cinema1, "Twilight", 17):
    cinema1 = vendebilhete(cinema1, "Twilight", 17)

listardisponibilidades(cinema1)
menu()
```
