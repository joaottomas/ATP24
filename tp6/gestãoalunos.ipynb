```python
def criar_turma():
    return  []


def inserir_aluno(turma):
        nome = input("Nome do aluno: ")
        id_aluno = input("ID do aluno: ")
        notaTPC = float(input("Nota TPC: "))
        notaProj = float(input("Nota Projeto: "))
        notaTeste = float(input("Nota Teste: "))
        aluno = (nome, id_aluno, [notaTPC, notaProj, notaTeste])
        turma.append(aluno)
        print(f"\nAluno {nome} inserido com sucesso!")

def listar_turma(turma):
    if not turma:
        print("\nA turma está vazia.")
    else:
        for aluno in turma:
            nome, id_aluno, notas = aluno
            print(f"\nNome: {nome}, ID: {id_aluno}, Notas: TPC {notas[0]}, Projeto {notas[1]}, Teste {notas[2]}")
    

def consultar_aluno(turma, id_aluno):
    for aluno in turma:
        if aluno[1] == id_aluno:
            nome, id_aluno, notas = aluno
            print(f"\nNome: {nome}, ID: {id_aluno}, Notas: TPC {notas[0]}, Projeto {notas[1]}, Teste {notas[2]}")
            return
    print("\nAluno não encontrado.")


def guardar_turma_ficheiro(turma, turmanome):
    file = open(turmanome,"w")
    for aluno in turma:
        nome, id_aluno, notas = aluno
        linha = f"{nome};{id_aluno};{notas[0]};{notas[1]};{notas[2]}\n"
        file.write(linha)
    file.close()
    print(f"\nTurma guardada em {turmanome}.")

def carregar_turma_ficheiro(turmanome):
    turma = []
    file = open(turmanome, 'r')
    for linha in file:
            partes = linha.strip().split(';')
            nome = partes[0]
            id_aluno = partes[1]
            notas = [float(partes[2]), float(partes[3]), float(partes[4])]
            aluno = (nome, id_aluno, notas)
            turma.append(aluno)
    file.close()
    print(f"\nTurma carregada de {turmanome}.")
    return turma


def menu():
    turma = []
    continuar = True
    while continuar:
        print("\nMenu de Operações:")
        print("1: Criar uma turma")
        print("2: Inserir um aluno na turma")
        print("3: Listar a turma")
        print("4: Consultar um aluno por ID")
        print("8: Guardar a turma em ficheiro")
        print("9: Carregar uma turma de um ficheiro")
        print("0: Sair da aplicação")
        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            turma = criar_turma()
            print("\nTurma criada com sucesso.")
        elif opcao == '2':
            inserir_aluno(turma)
        elif opcao == '3':
            listar_turma(turma)
        elif opcao == '4':
            id_aluno = input("ID do aluno: ")
            consultar_aluno(turma, id_aluno)
        elif opcao == '8':
            turmanome = input("Nome do ficheiro para guardar a turma: ")
            guardar_turma_ficheiro(turma, turmanome)
        elif opcao == '9':
            turmanome = input("Nome do ficheiro para carregar a turma: ")
            turma = carregar_turma_ficheiro(turmanome)
        elif opcao == '0':
            print("\nSaindo da aplicação.")
            continuar = False
        else:
            print("\nOpção inválida, tente novamente.")
menu()
```
